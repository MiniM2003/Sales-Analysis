```python
import pandas as pd
```

```python
df = pd.read_csv("Superstore.csv" , encoding="latin1")
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.duplicated().sum()
```

```python
df["Order Date"] = pd.to_datetime(df['Order Date'] , format="%d-%m-%Y")

df["Ship Date"] = pd.to_datetime(df['Ship Date'] , format="%d-%m-%Y")
```

```python
df.info()
```

```python
df['Order Date']
```

```python
df['Ship Date']
```

```python
df["Profit Margin %"] = (df['Profit'] / df['Sales']) * 100 
```

```python
df["Profit Margin %"]
```

```python
df['Year'] = df["Order Date"].dt.year
df['Year']
```

```python
df['Month'] = df["Order Date"].dt.month
df['Month']
```

```python
df['Quarter'] = df['Order Date'].dt.quarter
df['Quarter'] 
```

```python
df['Shipping Days'] = (df['Ship Date']- df['Order Date']).dt.days
df['Shipping Days']
```

```python
# Does increasing discount reduce profit?
# Average profit based on the discount
df.groupby('Discount')['Profit'].mean()
```

```python
df['Sales Category'] = pd.cut(
    df['Sales'],
    bins=[0,100,500,1000,5000,25000],
    labels=[
        'Very Low',
        'Low',
        'Medium',
        'High',
        'Very High'
    ]
)
```

```python
df['Sales Category'].head(10)
```

```python
# Total Sales
Total_sales = df['Sales'].sum()
Total_sales
```

```python
Total_profit = df["Profit"].sum()
print(Total_profit)
```

```python
# Average of Sales based on Region wise
df.groupby('Region')['Sales'].mean()
```

```python
# Region-wise average of Sales and Profit
df.groupby('Region')[['Sales','Profit']].mean()
```

```python
# Region and City-wise average sales
df.groupby(['Region', 'City'])['Sales'].mean()
```

```python
# Average Profit for each product Category
df.groupby('Category')['Profit'].mean()
```

```python
df.groupby('Region')[['Quantity','Sales','Profit']].mean()
```

## Visualization

```python
import matplotlib.pyplot as plt
```

```python
# Sales by Category
df.groupby('Category')['Sales'].sum().plot(kind='pie' , ylabel='')

plt.title("Sales by Category")
plt.show()
```

```python
# Sales by year
df.groupby('Year')['Sales'].sum().plot(kind='bar')

plt.title("Sales by Year")
plt.xlabel("Year")
plt.ylabel("Total Sales")
plt.show()
```

```python
# Monthly Sales Trend
df.groupby('Month')['Sales'].sum().plot(kind='line')

plt.title("Monthly trends")
```

```python
# Region Wise Profit
df.groupby('Region')['Profit'].sum().plot(kind='bar')
```

```python
import seaborn as sns

sns.heatmap(
df[['Sales','Profit','Quantity','Discount']].corr(),
annot=True
)
```
